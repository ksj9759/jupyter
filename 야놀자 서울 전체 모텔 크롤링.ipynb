```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.yanolja.com/motel?keyword=%EC%84%9C%EC%9A%B8&searchType=region'
motel_list = []

for page in range(1, 11):
    response = requests.get(f'{url}&page={page}')
    soup = BeautifulSoup(response.text, 'html.parser')

    motels = soup.select('.hotel_item')
    for motel in motels:
        name = motel.select_one('.name').get_text(strip=True)
        address = motel.select_one('.address').get_text(strip=True)
        score = motel.select_one('.score').get_text(strip=True)
        review_count = motel.select_one('.review_count').get_text(strip=True)
        price = motel.select_one('.price').get_text(strip=True)

        motel_info = {
            'name': name,
            'address': address,
            'score': score,
            'review_count': review_count,
            'price': price
        }
        motel_list.append(motel_info)

for motel in motel_list:
    print(motel)
```
