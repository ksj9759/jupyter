```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
url = 'https://www.yanolja.com'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)

# 모델버튼 클릭
driver.find_element(By.CSS_SELECTOR, '#__next > div.StyleComponent_container__1jS9A.index_container__3kY3K > section.index_noTabContentWrapper__1zz-v > section > section > section:nth-child(1) > div:nth-child(1) > a:nth-child(1) > p').click()

# 서울 버튼 클릭후 soup 다시 가져오기
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.find_element(By.CSS_SELECTOR, '#__next > div:nth-child(1) > div.StyleComponent_container__1jS9A.SubhomeBody_layout__3oFMz > main > div.SubhomeRegionList_regionListWrap__3VC7Z > ul > li:nth-child(1) > a').click()
seoul_button_click = BeautifulSoup(driver.page_source, 'html.parser')

# 21개 리스트에서 강남권 클릭
href = soup.select('div.SubhomeRegionList_regionListWrap__3VC7Z > ul > li.SubhomeRegionList_regionActive__39Bnt > div > a')
driver.find_element(By.CSS_SELECTOR, '#__next > div:nth-child(1) > div.StyleComponent_container__1jS9A.SubhomeBody_layout__3oFMz > main > div.SubhomeRegionList_regionListWrap__3VC7Z > ul > li.SubhomeRegionList_regionActive__39Bnt > div > a').click()
Seoul_specific = BeautifulSoup(driver.page_source, 'html.parser')

# 야놀자 추천순 클릭
driver.find_element(By.CSS_SELECTOR, '#__next > div.StyleComponent_container__1jS9A.list_listContainer__2kL99.list_bottomPadding__xvWzu > section.PlaceListSortNav_container__1MtJz > button').click()
recommend = BeautifulSoup(driver.page_source, 'html.parser')
time.sleep(2)
# 찜많은 순 클릭
driver.find_element(By.CSS_SELECTOR, 'body > div:nth-child(43) > div > div > section > div > ul > li:nth-child(6) > button').click()
time.sleep(2)
popular = BeautifulSoup(driver.page_source, 'html.parser')

# 강남/역삼/삼성/논현 숙박업 리스트
app = popular.select('div.PlaceListItemText_area__3l67D')

lines = []
for li in app:
    
    # 업소명 , 업소 대실 가격, 업소 숙박 가격, 대실 시간
    title = li.find('strong').get_text().strip()
    print(title)
    price_room = li.select('span.PlacePriceInfo_salePrice__28VZD')[0].text
    price_accomodation = li.select('span.PlacePriceInfo_salePrice__28VZD')[1].get_text()
    lent = li.select_one('span.PlacePriceInfo_extraInfoText__1bCwu').get_text()
    time.sleep(2)
    
    # 상세페이지 클릭
    driver.find_element(By.CSS_SELECTOR, '#__next > div.StyleComponent_container__1jS9A.list_listContainer__2kL99.list_bottomPadding__xvWzu > section.PlaceListBody_placeListBodyContainer__1u70R.PlaceListBody_hideHeader__1iioo > div > div > div:nth-child(1) > a > div.PlaceListItemText_area__3l67D > div.PlaceListItemText_image__2mlnK.place-image > div.PlaceListImage_container__t-mGF > div.PlaceListImage_imageText__2XEMn').click()
    # 인기시설 및 서비스 가져오기
    popular_service = BeautifulSoup(driver.page_source, 'html.parser')
    service = popular_service.select('div.css-9cehu5')
    # 위치/정보 클릭후 도로명 주소 가져오기
    driver.find_element(By.CSS_SELECTOR, '#__next > div > div > main > article > div:nth-child(2) > div > div.css-1urbzj6 > div > button:nth-child(2)').click()
    road_juso = BeautifulSoup(driver.page_source, 'html.parser')
    location = road_juso.select_one('div.css-o8j33g').text.strip('주소복사')
    # 시설/서비스 정보 클릭후 객실 수 가져오기
    driver.find_element(By.CSS_SELECTOR, '#__next > div > div > main > article > div:nth-child(2) > div > div.css-1urbzj6 > div > button:nth-child(4)').click()
    facility_page =BeautifulSoup(driver.page_source, 'html.parser')
    room_num = facility_page.select_one('div.description').get_text().strip('개')
    # 후기 클릭후 댓글 정보 가져오기
    driver.find_element(By.CSS_SELECTOR, '#__next > div > div > main > article > div:nth-child(2) > div > div.css-1urbzj6 > div > button:nth-child(5)').click()
    review_page =BeautifulSoup(driver.page_source, 'html.parser')
    # 스크롤 2페이지 내린후 후기 정보 가져오기
    last_height = driver.execute_script("return document.body.scrollHeight")
    for i in range(2):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
    
    review_stroy_page = BeautifulSoup(driver.page_source, 'html.parser')    
    review_stroy = review_stroy_page.select('div.css-1byy3oj')
    lines.append([title, price_room, price_accomodation, lent,service,location,room_num,review_stroy])
    
    driver.back()
```
